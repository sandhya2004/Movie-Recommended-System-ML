```python
#conert string of list into list with the help of ast
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#The above formate of genres is wiered we have to change  it foramte
# To convert above formate into ['Action' , 'Adeventure' , 'Fantasy' , 'Scifi']
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python






#movies['cast'][0] # It have multiple charcter we need from straing only 3 or 4 char
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter <=3:
            L.append(i['name'])
        else:
            break
        
    return L    
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```
